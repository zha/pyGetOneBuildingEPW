```python
import requests
import re
```

```python
LocationPatten = "href=\\\"([a-zA-Z0-9-._/]+?)\\.html";
zipFilePatten = "href=\\\"([a-zA-Z0-9-._/]+?)\\.zip";
link = "http://climate.onebuilding.org/";

region_link = 'http://climate.onebuilding.org/WMO_Region_4_North_and_Central_America/'

region_request = requests.get('http://climate.onebuilding.org/WMO_Region_4_North_and_Central_America/', allow_redirects=False)

locs_list = re.findall(LocationPatten, region_request.content.decode("utf-8") )

index_file = open("index.txt", 'w')
for loc in locs_list:
    zipstr = requests.get('http://climate.onebuilding.org/WMO_Region_4_North_and_Central_America/' + loc.replace("index", ""), allow_redirects=False)
    zips_list = re.findall(zipFilePatten, zipstr.content.decode("utf-8") )
    
    for ind_zip in zips_list:
        print(region_link + loc.replace("index", "") + ind_zip)
        index_file.write(region_link + loc.replace("index", "") + ind_zip )
        
        
        r = requests.get(region_link + loc.replace("index", "") + ind_zip + '.zip')
        with open(ind_zip.split('/')[-1] + '.zip', 'wb') as f:
            f.write(r.content)
        
index_file.close()

    
```
